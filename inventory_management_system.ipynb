```python
from datetime import datetime
from tabulate import tabulate

INVENTORY_FILE = "inventory.txt"
SALES_FILE = "sales.txt"


def load_inventory():
    with open(INVENTORY_FILE, "r") as f:
        return [line.strip().split(",") for line in f if line.strip()]


def save_inventory(products):
    with open(INVENTORY_FILE, "w") as f:
        for prod in products:
            f.write(",".join(prod) + "\n")


def view_products():
    products = load_inventory()
    headers = ["ID", "Name", "Price", "Quantity"]
    print(tabulate(products, headers=headers, tablefmt="fancy_grid"))


def search_product():
    pid = input("Enter Product ID: ").strip()
    products = load_inventory()
    for prod in products:
        if prod[0] == pid:
            print(tabulate([prod], headers=["ID", "Name", "Price", "Quantity"], tablefmt="fancy_grid"))
            return
    print("Product not found.")


def purchase_product():
    products = load_inventory()
    pid = input("Enter Product ID: ").strip()
    qty_req = int(input("Enter quantity: ").strip())

    for prod in products:
        if prod[0] == pid:
            name, price, stock = prod[1], float(prod[2]), int(prod[3])
            print(f"\nProduct: {name}")
            print(f"Price: {price}")
            print(f"Available: {stock}")

            if qty_req > stock:
                print(f" Out of Stock! Only {stock} available.")
                return

            total = price * qty_req
            print(f" Total Amount: {total}")

            # Update stock
            prod[3] = str(stock - qty_req)
            save_inventory(products)

            # Save sale
            with open(SALES_FILE, "a") as f:
                f.write(f"{datetime.now()},{pid},{name},{qty_req},{price},{total}\n")

            print(f"Stock updated successfully. Now available: {stock - qty_req}")
            if stock - qty_req < 5:
                print("⚠️ Low stock! Consider restocking soon.")
            return
    print("Invalid Product ID.")


def add_product():
    products = load_inventory()
    pid = input("Enter New Product ID: ").strip()
    if any(prod[0] == pid for prod in products):
        print("Product ID already exists!")
        return
    name = input("Enter Product Name: ").strip()
    price = input("Enter Price: ").strip()
    qty = input("Enter Quantity: ").strip()
    products.append([pid, name, price, qty])
    save_inventory(products)
    print("Product added successfully.")


def update_product():
    products = load_inventory()
    pid = input("Enter Product ID to update: ").strip()
    for prod in products:
        if prod[0] == pid:
            prod[1] = input(f"Enter new name ({prod[1]}): ") or prod[1]
            prod[2] = input(f"Enter new price ({prod[2]}): ") or prod[2]
            prod[3] = input(f"Enter new quantity ({prod[3]}): ") or prod[3]
            save_inventory(products)
            print("Product updated successfully.")
            return
    print("Product not found.")


def delete_product():
    products = load_inventory()
    pid = input("Enter Product ID to delete: ").strip()
    products = [prod for prod in products if prod[0] != pid]
    save_inventory(products)
    print("Product deleted successfully.")


def sales_report():
    try:
        with open(SALES_FILE, "r") as f:
            sales = [line.strip().split(",") for line in f if line.strip()]
        headers = ["Date", "ID", "Name", "Qty", "Price", "Total"]
        print(tabulate(sales, headers=headers, tablefmt="fancy_grid"))
    except FileNotFoundError:
        print("No sales recorded yet.")


def main():
    while True:
        print("\n=== Inventory Management System ===")
        print("1. View All Products")
        print("2. Search Product")
        print("3. Purchase Product")
        print("4. Add New Product")
        print("5. Update Product")
        print("6. Delete Product")
        print("7. View Sales Report")
        print("8. Exit")

        choice = input("Enter choice: ").strip()

        if choice == "1":
            view_products()
        elif choice == "2":
            search_product()
        elif choice == "3":
            purchase_product()
        elif choice == "4":
            add_product()
        elif choice == "5":
            update_product()
        elif choice == "6":
            delete_product()
        elif choice == "7":
            sales_report()
        elif choice == "8":
            print("Exiting...")
            break
        else:
            print("Invalid choice.")

main()
```

```python
# from datetime import datetime
from datetime import datetime
from tabulate import tabulate
import shutil
import csv
import os

INVENTORY_FILE = "inventory.txt"
SALES_FILE = "sales.txt"
LOW_STOCK_THRESHOLD = 5


def load_inventory():
    with open(INVENTORY_FILE, "r") as f:
        return [line.strip().split(",") for line in f if line.strip()]


def save_inventory(products):
    backup_inventory()
    with open(INVENTORY_FILE, "w") as f:
        for prod in products:
            f.write(",".join(prod) + "\n")


def backup_inventory():
    if os.path.exists(INVENTORY_FILE):
        shutil.copy(INVENTORY_FILE, INVENTORY_FILE + ".bak")



def view_products():
    products = load_inventory()
    headers = ["ID", "Name", "Price", "Quantity", "Category"]
    print(tabulate(products, headers=headers, tablefmt="fancy_grid"))


def search_product():
    pid = input("Enter Product ID: ").strip()
    products = load_inventory()
    for prod in products:
        if prod[0] == pid:
            print(tabulate([prod], headers=["ID", "Name", "Price", "Quantity", "Category"], tablefmt="fancy_grid"))
            return
    print("Product not found.")


def low_stock_alert():
    products = load_inventory()
    low_stock_items = [p for p in products if int(p[3]) < LOW_STOCK_THRESHOLD]
    if low_stock_items:
        print("\n Low Stock Items:")
        print(tabulate(low_stock_items, headers=["ID", "Name", "Price", "Quantity", "Category"], tablefmt="fancy_grid"))
    else:
        print("\n All items have sufficient stock.")


def purchase_product():
    products = load_inventory()
    pid = input("Enter Product ID: ").strip()
    try:
        qty_req = int(input("Enter quantity: ").strip())
    except ValueError:
        print("Invalid quantity. Please enter a number.")
        return

    for prod in products:
        if prod[0] == pid:
            name, price, stock = prod[1], float(prod[2]), int(prod[3])
            print(f"\nProduct: {name}")
            print(f"Price: {price}")
            print(f"Available: {stock}")

            if qty_req > stock:
                print(f" Out of Stock! Only {stock} available.")
                return

            total = price * qty_req
            print(f"Total Amount: {total}")

            # Customer details
            cust_name = input("Enter Customer Name: ").strip()
            cust_contact = input("Enter Customer Contact Number: ").strip()
            cust_email = input("Enter Customer Email (optional): ").strip()

            # Update stock
            prod[3] = str(stock - qty_req)
            save_inventory(products)

            # Save sale
            with open(SALES_FILE, "a") as f:
                f.write(f"{datetime.now()},{pid},{name},{qty_req},{price},{total},{cust_name},{cust_contact},{cust_email}\n")

            # Print Bill
            print("\n" + "="*40)
            print("        ANUBHAV STORE - BILL RECEIPT")
            print("="*40)
            print(f"Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
            print(f"Customer: {cust_name}")
            print(f"Contact: {cust_contact}")
            if cust_email:
                print(f"Email: {cust_email}")
            print("-"*40)
            table = [[pid, name, price, qty_req, total]]
            print(tabulate(table, headers=["ID", "Name", "Price", "Qty", "Total"], tablefmt="fancy_grid"))
            print("-"*40)
            print(f"Grand Total: {total:.2f}")
            print("="*40)
            print("   Thank you for shopping with us !")
            print("="*40)

            if stock - qty_req < LOW_STOCK_THRESHOLD:
                print(" low stock! Consider restocking soon.")
            return
    print("Invalid Product ID.")

def add_product():
    products = load_inventory()
    pid = input("Enter New Product ID: ").strip()
    if any(prod[0] == pid for prod in products):
        print("Product ID already exists!")
        return
    name = input("Enter Product Name: ").strip()
    price = input("Enter Price: ").strip()
    qty = input("Enter Quantity: ").strip()
    category = input("Enter Category: ").strip()
    products.append([pid, name, price, qty, category])
    save_inventory(products)
    print("12Product added successfully.")


def update_product():
    products = load_inventory()
    pid = input("Enter Product ID to update: ").strip()
    for prod in products:
        if prod[0] == pid:
            prod[1] = input(f"Enter new name ({prod[1]}): ") or prod[1]
            prod[2] = input(f"Enter new price ({prod[2]}): ") or prod[2]
            prod[3] = input(f"Enter new quantity ({prod[3]}): ") or prod[3]
            prod[4] = input(f"Enter new category ({prod[4]}): ") or prod[4]
            save_inventory(products)
            print("Product updated successfully.")
            return
    print("Product not found.")


def delete_product():
    products = load_inventory()
    pid = input("Enter Product ID to delete: ").strip()
    products = [prod for prod in products if prod[0] != pid]
    save_inventory(products)
    print("Product deleted successfully.")


def restock_product():
    products = load_inventory()
    pid = input("Enter Product ID to restock: ").strip()
    for prod in products:
        if prod[0] == pid:
            try:
                add_qty = int(input("Enter quantity to add: ").strip())
            except ValueError:
                print("Invalid quantity.")
                return
            prod[3] = str(int(prod[3]) + add_qty)
            save_inventory(products)
            print(f"Restocked successfully. New quantity: {prod[3]}")
            return
    print("Product not found.")



def sales_report():
    try:
        with open(SALES_FILE, "r") as f:
            sales = [line.strip().split(",") for line in f if line.strip()]
        headers = ["Date", "ID", "Name", "Qty", "Price", "Total"]
        print(tabulate(sales, headers=headers, tablefmt="fancy_grid"))
    except FileNotFoundError:
        print("No sales recorded yet.")


def export_sales_to_csv():
    try:
        with open(SALES_FILE, "r") as f:
            sales = [line.strip().split(",") for line in f if line.strip()]
        with open("sales_export.csv", "w", newline="") as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(["Date", "ID", "Name", "Qty", "Price", "Total"])
            writer.writerows(sales)
        print("Sales exported to sales_export.csv")
    except FileNotFoundError:
        print("No sales data to export.")


def sales_summary():
    try:
        with open(SALES_FILE, "r") as f:
            sales = [line.strip().split(",") for line in f if line.strip()]
        total_revenue = sum(float(s[5]) for s in sales)
        total_items = sum(int(s[3]) for s in sales)
        print(f" Total Revenue: {total_revenue}")
        print(f" Total Items Sold: {total_items}")
    except FileNotFoundError:
        print("No sales recorded yet.")



def main():
    while True:
        print("\n--- Inventory Management System ---")
        print("1. View All Products")
        print("2. Search Product")
        print("3. Purchase Product")
        print("4. Add New Product")
        print("5. Update Product")
        print("6. Delete Product")
        print("7. View Sales Report")
        print("8. Low Stock Alert")
        print("9. Restock Product")
        print("10. Export Sales to CSV")
        print("11. Sales Summary")
        print("12. Exit")

        choice = input("Enter choice: ").strip()

        if choice == "1":
            view_products()
        elif choice == "2":
            search_product()
        elif choice == "3":
            purchase_product()
        elif choice == "4":
            add_product()
        elif choice == "5":
            update_product()
        elif choice == "6":
            delete_product()
        elif choice == "7":
            sales_report()
        elif choice == "8":
            low_stock_alert()
        elif choice == "9":
            restock_product()
        elif choice == "10":
            export_sales_to_csv()
        elif choice == "11":
            sales_summary()
        elif choice == "12":
            print("Exiting...")
            break
        else:
            print("Invalid choice.")


main()
```
